## 先複習類別 ( class )
在python中，使用關鍵字，class來定義類別。
類別的格式(雛形)如下所示。<br><br>
class 類別名稱：<br>
    def __ init __ (self, 參數, ...): #建構子<br>
        ...<br>
    def 方法名稱1 (self, 參數, ...): #方法一<br>
        ...<br>
    def 方法名稱2 (self, 參數, ...): #方法二<br>
        ...<br>
這裡出現 __ init __這種特別的方法，這是執行初始化的方法。<br>
初始化的方法也稱為 建構子( constructor )，建立類別( class )的實體時，只會呼叫一次。<br>
另外，Python的特色是在方法的第一參數之中，清楚地寫出代表本身( 實體 )的self。<br><br>



```python
#接著，我們來實作看看
class Man:
    def __init__(self, name):
        self.name = name
        print("Initilized!")
    
    def hello(self):
        print("Hello" + self.name + "!")
    
    def goobye(self):
        print("Goodbye" + self.name + "!")
```

```python
m = Man("David")
m.hello()
m.goobye()
```

這裡，我們定義了新類別。在上面的例子中，從Man這個類別產生了稱作m的實體( 物件 )。<br><br>
Man類別的建構子 ( 初始化方法 ) 是取得name參數，利用這個參數，將實例變數self.name 初始化。<br>
「*實例變數*」 是指，儲存在各個實體的變數。在python中，可以像self.name一樣，在self之後輸入屬性名稱，即可建立並存取實例變數。


## Numpy
執行深度學習時，常需要計算陣列或矩陣。Numpy的陣列類別( numpy.array )提供許多方便的方法。<br>
深度學習也會使用這些方法。以下將會簡介後續會用到的Numpy。


```python
#Numpy 陣列的算術運算
import numpy as np
x = np.array([1, 2 ,3])
y = np.array([2, 4 ,6])
print(x+y)
print(x-y)
print(x*y) #分量橫向相乘
print(x/y)

#Numpy 也能進行廣播( broadcast )運算
x = np.array([1, 4, 6])
x/2
```

```python
#Numpy 的N維陣列
#Numpy 不僅能建立一維陣列，也能建立多為陣列，如下。
A = np.array([[1,2],[3,4]])
print(A)
print(A.shape)
print(A.dtype)
```

```python
B = np.array([[3,0],[0,6]])
print(A+B)
print(A*B) #內積
```

## 廣播( broadcast )
Numpy也可以進行不同形狀的陣列運算，其中之一為廣播，如圖所示。<br>

## Matplotlib
在深度學習中，繪製圖表與資料視覺會變得更重要。Matplotlib是繪製圖表用的函式庫。<br>
使用Matplotlib，可以輕鬆第繪製圖表或將資料視覺化。以下將說明其如何處理資料將其視覺化。


```python
# 繪製單純的圖表
# 若要繪製圖表，我們就要利用matplotlib中的pyplot模組
# 讓我們來看看sin函數長甚麼樣子吧
import numpy as np
import matplotlib.pyplot as plt

# 建立資料
x = np.arange(0, 6, 0.02)
y = np.sin(x)

# 繪製圖表
plt.plot(x, y)
plt.show()
```

```python
# pyplot的功能
# 除了介紹前面的sin函數，我們也使用cos來繪圖
# 並使用其他pyplot的功能，加上標題與x軸與其標籤的名稱。

# 建立資料
x = np.arange(0, 6, 0.2)
y1 = np.sin(x)
y2 = np.cos(x)

# 繪製圖表
plt.plot(x, y1, label="sin")
plt.plot(x, y2, linestyle = '--', label = 'cos') # 用虛線繪圖
plt.xlabel('x') # x軸的標籤
plt.ylabel('y')
plt.title('sin & cos') # 標題
plt.legend() # 圖例( legend )旨在説明我們顯示與每個資料相對應的圖像名稱。讓讀者更好地瞭解您的資料結構。
plt.show()
```

```python
# 顯示圖像
# pyplot也提供顯示影像用的方法，imshow()。
# 另外，如果要載入影像，可以利用plt.image模組的imread()。

import matplotlib.pyplot as plt
from matplotlib.image import imread

img = imread('broadcast.jpg') # 載入圖像(設定適當的路徑!)
plt.imshow(img)

plt.show()
```
